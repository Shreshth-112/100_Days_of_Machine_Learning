```python
import matplotlib.pyplot as plt
import numpy as np

# Original tensor
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Transposed tensor
A_T = A.T

# Plotting the original tensor
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(A, cmap='viridis', interpolation='none')
plt.title('Original Tensor')
plt.colorbar()

# Plotting the transposed tensor
plt.subplot(1, 2, 2)
plt.imshow(A_T, cmap='viridis', interpolation='none')
plt.title('Transposed Tensor')
plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Original tensor
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Transposed tensor
A_T = A.T

# Function to plot vectors
def plot_vectors(tensor, title, subplot):
    rows, cols = tensor.shape
    for i in range(rows):
        for j in range(cols):
            plt.arrow(0, 0, j, i, head_width=0.1, head_length=0.1, fc='blue', ec='blue')
            plt.text(j, i, f'{tensor[i, j]}', fontsize=12, ha='right')
    plt.xlim(-1, cols)
    plt.ylim(-1, rows)
    plt.title(title)
    plt.gca().invert_yaxis()
    plt.grid()

# Plotting the original tensor
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plot_vectors(A, 'Original Tensor', 1)

# Plotting the transposed tensor
plt.subplot(1, 2, 2)
plot_vectors(A_T, 'Transposed Tensor', 2)

plt.show()
```
